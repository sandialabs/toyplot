```python
import arrow
import datetime
import numpy
import toyplot.data
import toyplot.locator

data_table = toyplot.data.read_csv("../docs/temperatures.csv")
data_table["DATE"] = [arrow.get(value, "YYYYMMDD") for value in data_table["DATE"]]
data_table["TIMESTAMP"] = [value.timestamp for value in data_table["DATE"]]
data_table["TMAX"] = data_table["TMAX"].astype("float64") * 0.1
data_table["TMIN"] = data_table["TMIN"].astype("float64") * 0.1
data_table["TOBS"] = data_table["TOBS"].astype("float64") * 0.1

canvas = toyplot.Canvas(width=800, height=300)
axes = canvas.axes(ylabel=u"Temperature \u00b0C")
axes.x.ticks.locator = toyplot.locator.Timestamp()
axes.x.ticks.show = True
axes.plot(data_table["TIMESTAMP"], data_table["TMAX"], color="red", stroke_width=1)
axes.plot(data_table["TIMESTAMP"], data_table["TMIN"], color="blue", stroke_width=1);
```

```python
begin = arrow.get(2016, 11, 6, 0).timestamp
end = arrow.get(2016, 11, 7, 0, 0, 0).timestamp

numpy.random.seed(1234)
events = numpy.random.uniform(begin, end, size=100)
canvas = toyplot.Canvas(width=800, height=200)
axes = canvas.number_line(min=begin, max=end)
axes.axis.ticks.locator = toyplot.locator.Timestamp(interval=(5, "hours"))
axes.axis.ticks.show = True
axes.scatterplot(events, marker="|", offset=20, size=200)
```
