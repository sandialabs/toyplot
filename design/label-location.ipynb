```python
import numpy
import toyplot
```

```python
canvas = toyplot.Canvas(width=600, height=300)
numberline = canvas.numberline(label="The Label")
numberline.axis.ticks.location = "above"
numberline.axis.label.location = "above"
```

```python
canvas = toyplot.Canvas(width=600, height=600)
axes = canvas.cartesian(xlabel="X Label", ylabel="Y Label", label="Label")
axes.x.label.style={"font-size":"16pt"}
axes.y.label.style={"font-size":"16pt"}
axes.y.spine.position="high"
```
