```python
import numpy
import toyplot
import toyplot.color
```

```python
pl = toyplot.color.Palette()
x = numpy.linspace(0, 1)
y1 = (1 - x) ** 2
y2 = numpy.column_stack((1 - (x ** 2), x ** 2))

canvas = toyplot.Canvas(width=600, height=300)
m1 = canvas.axes(grid=(1,2,0), gutter=15).scatterplot(x, y1, marker="o", color=pl[0])
m2 = canvas.axes(grid=(1,2,1), gutter=15, yshow=False).scatterplot(x, y2, marker="s", color=[pl[1], pl[2]])

canvas.legend([
    ("Experiment 1", "o", {"fill":pl.css(0), "stroke": "none"}),
    ("Experiment 2", "s", {"fill":pl.css(1), "stroke": "none"}),
    ("Experiment 3", "s", {"fill":pl.css(2), "stroke": "none"}),],
    corner=("top", 100, 100, 70),
    );
```
