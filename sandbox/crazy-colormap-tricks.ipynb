```python
import numpy
import toyplot

numpy.random.seed(1234)
```

```python
colormap1 = toyplot.color.LinearMap(toyplot.color.Palette([toyplot.color.css("red"), toyplot.color.rgba(1, 0, 0, 0)]), domain_min=0, domain_max=1)
colormap1
```

A custom colormap from opaque-blue to transparent:

```python
colormap2 = toyplot.color.LinearMap(toyplot.color.Palette([toyplot.color.css("blue"), toyplot.color.rgba(0, 0, 1, 0)]), domain_min=0, domain_max=1)
colormap2
```

```python
data1 = numpy.random.uniform(size=(4, 4))
data2 = numpy.random.uniform(size=(4, 4))

toyplot.matrix((data1, colormap1))

toyplot.matrix((data2, colormap2))

canvas = toyplot.Canvas()
canvas.matrix((data1, colormap1))
canvas.matrix((data2, colormap2), tshow=False, lshow=False); # Don't keep overwriting the same labels
```
