```python
import numpy
import toyplot

toyplot.log.setLevel(toyplot.logging.DEBUG)
```

```python
series = numpy.column_stack((numpy.linspace(0, 1), numpy.linspace(0, 1)**2))

legend = []
canvas = toyplot.Canvas(width=400)
axes = canvas.cartesian()
legend.append(("A", axes.plot(series)))
legend.append(("B", axes.plot(numpy.linspace(0, 1)**3, style={"stroke-dasharray":"2,2"})))
canvas.legend(legend, corner=("top-left", 50, 100, 50), style={"fill":"#f4f4f4"});
```

```python
series = numpy.column_stack((numpy.linspace(0, 1), numpy.linspace(0, 1)**2))

legend = []
canvas = toyplot.Canvas(width=400)
axes = canvas.cartesian()
legend.append(("A", axes.bars(series, baseline="symmetric")))
legend.append(("B", axes.bars(numpy.linspace(0, 1)**3)))
canvas.legend(legend, corner=("top-left", 50, 100, 50), style={"fill":"#f4f4f4"});
```

```python
series = numpy.column_stack((numpy.linspace(0, 1), numpy.linspace(0, 1)**2))

legend = []
canvas = toyplot.Canvas(width=400)
axes = canvas.cartesian()
legend.append(("A", axes.fill(series, baseline="symmetric")))
legend.append(("B", axes.fill(numpy.linspace(0, 1)**3, style={"stroke":"black"})))
canvas.legend(legend, corner=("top-left", 50, 100, 50), style={"fill":"#f4f4f4"});
```

```python
series = numpy.column_stack((numpy.linspace(0, 1), numpy.linspace(0, 1)**2))

legend = []
canvas = toyplot.Canvas(width=400)
axes = canvas.cartesian()
legend.append(("A", axes.scatterplot(series)))
legend.append(("B", axes.scatterplot(numpy.linspace(0, 1)**3, color="red", marker={"shape":"d", "label":"1"}, size=15, mstyle={"stroke":"black"}, mlstyle={"fill":"white"})))
canvas.legend(legend, corner=("top-left", 50, 100, 60), style={"fill":"#f4f4f4"});
```
